```python
from bs4 import BeautifulSoup
import pandas as pd
import time
from time import sleep
from random import randint
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('http://www.chainplay.gg/gaming-ico/')
```

```python
name=[]
current_price=[]
sale_price=[]
market_cap=[]
raised=[]
roi_usd=[]
ath_roi=[]
type=[]
tge_platform=[]
ended=[]


elements=driver.find_elements(By.CSS_SELECTOR, 'table tbody a[class="text-3.25 sm:text-sm font-medium"]')
for element in range(len(elements)):
    name.append(elements[element].text)


elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(2)')
for element in range(len(elements)):
    current_price.append(elements[element].text)

    
elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(3)')
for element in range(len(elements)):
    sale_price.append(elements[element].text)   


elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(4)')
for element in range(len(elements)):
    market_cap.append(elements[element].text)    


elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(5)')
for element in range(len(elements)):
    raised.append(elements[element].text)


elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(6)')
for element in range(len(elements)):
    roi_usd.append(elements[element].text)

    
elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(7)')
for element in range(len(elements)):
    ath_roi.append(elements[element].text)   


elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(8)')
for element in range(len(elements)):
    type.append(elements[element].text)


elements=driver.find_elements(By.CSS_SELECTOR, 'a[class="font-medium text-sm capitalize"]')
for element in range(len(elements)):
    tge_platform.append(elements[element].text)    
    

elements=driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(10)')
for element in range(len(elements)):
    ended.append(elements[element].text)
    

print(name)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(current_price)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(sale_price)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(market_cap)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(raised)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(roi_usd)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(ath_roi)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(type)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(tge_platform)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(ended)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(len(name),len(current_price),len(sale_price),len(market_cap),len(raised),len(roi_usd),len(ath_roi),len(type),len(tge_platform),len(ended))
```
