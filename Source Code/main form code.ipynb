```python
from bs4 import BeautifulSoup
import pandas as pd
import time
from time import sleep
from random import randint
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())
driver.maximize_window()
driver.implicitly_wait(2) 
driver.get("https://chainplay.gg")
```

```python
BigName=[]
tokenname=[]
rowlink=[]
genre=[]
platform=[]
tokenprice=[]
updown=[]
tokenvolume=[]
marketcap=[]
infourl=[]
#爬游戏名字
elements = driver.find_elements(By.CSS_SELECTOR, 'table tbody tr[class="border-b hover:bg-gray-100 "] div[class="flex-1 overflow-hidden"] p')
for a in range(len(elements)):
    BigName.append(elements[a].text)

#爬游戏详情页链接
elements = driver.find_elements(By.CSS_SELECTOR,'a[class="flex flex-col lg:flex-row md:flex-row items-center "]')
for element in elements:
    infourl.append(element.get_attribute('href'))

#爬游戏代币名称
elements = driver.find_elements(By.CSS_SELECTOR, 'div[class="flex-1 overflow-hidden"] small')
for b in range(len(elements)):
    tokenname.append(elements[b].text)


#爬区块链的图标链接
page=driver.page_source
soup=BeautifulSoup(page,"lxml")
b=soup.find_all('td',attrs={'class':'px-1.5 sm:px-3.75 py-2.5 text-right'})
for data in b:
    j=[]
    for x in data.find_all('a'):
        j.append(x['href'])
    rowlink.append(j)

#爬游戏类型
elements = driver.find_elements(By.CSS_SELECTOR, 'td[class="px-2.5 sm:px-3.75 py-2.5 text-right text-3.25 capitalize"]')
for b in range(len(elements)):
    genre.append(elements[b].text)

#爬游戏运行平台
page=driver.page_source
soup=BeautifulSoup(page,"lxml")
b=soup.select('td[class="px-2.5 sm:px-3.75 py-2.5 text-right"] > div')
for data in b:
    j=[]
    for i in data.find_all('a'):
        j.append(i['href'])
    platform.append(j)   

#爬代币价格
elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100 "] td:nth-of-type(6)')
for element in elements:
    tokenprice.append(element.text)

#代币价格24h涨跌
elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100 "] td:nth-of-type(7)')
for element in elements:
    updown.append(element.text)

#代币价值
elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100 "] td:nth-of-type(8)')
for element in elements:
    tokenvolume.append(element.text)

#游戏市值
elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100 "] td:nth-of-type(9)')
for element in elements:
    marketcap.append(element.text)

print(BigName[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(infourl[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(tokenname[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(rowlink[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(genre[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(platform[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(tokenprice[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(updown[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(tokenvolume[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(marketcap[0])
print('------------------------------------------------------------------------------------------------------------------------------------')
print(len(BigName),len(infourl),len(tokenname),len(rowlink),len(genre),len(platform),len(tokenprice),len(updown),len(tokenvolume),len(marketcap))
```

```python
#清理获取链接的无用字符
blockchain=[]
block1=[]
for a in rowlink:
    inner=[]
    for b in a:
        inner.append(b.replace('/chain/','').replace('/',''))
    blockchain.append(inner)
for a in blockchain:
    b=",".join(a)
    block1.append(b)
print(block1)
```

```python
platformclean=[]
plat1=[]
for a in platform:
    inner=[]
    for b in a:
        inner.append(b.replace('/platform/','').replace('/',''))
    platformclean.append(inner)
for element in platformclean:
    a=','.join(element)
    plat1.append(a)
print(plat1)
```

```python
genreclean=[]
for a in genre:
    genreclean.append(a.replace("\n",", "))
len(genreclean)
```

```python
import random
tokencontract=[]
scholarguild=[]
partneredgamenumber=[]

for i in infourl:
    driver.get(i)
    element = driver.find_element(By.CSS_SELECTOR, 'button[class="group  uppercase "]:nth-of-type(2)')
    element.click()
    time.sleep(0.2)
    element = driver.find_element(By.CSS_SELECTOR, 'div[class="bg-gray-100 rounded-xl flex flex-row items-center px-3 py-[2px] "] button')
    element.click()
    data=pd.read_clipboard()
    data.columns.to_list()
    a= ''.join(data.columns.to_list())
    tokencontract.append(a)
    
    element = driver.find_element(By.CSS_SELECTOR, 'button[class="group  uppercase "]:nth-last-child(1)')
    element.click()
    time.sleep(0.2)
    elements=driver.find_elements(By.CSS_SELECTOR, 'tr td[class=" sticky px-2.5 sm:px-3.75 bg-white py-5 text-left sm:max-w-[13.75rem] sm:w-[13.75rem]"]')
    scholarguildin=[]
    for element in elements:
        scholarguildin.append(element.text)
    partneredgamenumberin=[]
    elements = driver.find_elements(By.CSS_SELECTOR, 'td[class="px-2.5 sm:px-3.75 py-2.5 text-right"]:nth-child(7)')
    for element in elements:
        partneredgamenumberin.append(element.text)
    
    scholarguild.append(scholarguildin)
    partneredgamenumber.append(partneredgamenumberin)    
print(len(tokencontract))
print(len(scholarguild))
```
