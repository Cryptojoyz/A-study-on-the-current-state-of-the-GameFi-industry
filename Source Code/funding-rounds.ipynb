```python
from bs4 import BeautifulSoup
import pandas as pd
import time
from time import sleep
from random import randint
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('http://www.chainplay.gg/funding-rounds/')
```

```python
project=[]
funding_rounds=[]
total_funding_amount=[]
investors=[]
date=[]


elements= driver.find_elements(By.CSS_SELECTOR, 'table tbody div[class="flex-1 overflow-hidden"] a')
for element in range(len(elements)):
    project.append(elements[element].text)


elements=driver.find_elements(By.CSS_SELECTOR, 'td[class="px-2.5 sm:px-3.75 py-2.5 text-left"]')
for element in range(len(elements)):
    funding_rounds.append(elements[element].text)

    
elements=driver.find_elements(By.CSS_SELECTOR, 'td[class="px-2.5 sm:px-3.75 py-2.5 text-right"]')
for element in range(len(elements)):
    total_funding_amount.append(elements[element].text)


elements=driver.find_elements(By.CSS_SELECTOR,'tr[class="py-1"] td:nth-of-type(4) button ')
for element in elements:
    element.click()

elements=driver.find_elements(By.CSS_SELECTOR, 'td[class="px-2.5 sm:px-3.75 py-2.5"]')
for element in range(len(elements)):
    investors.append(elements[element].text)
       
    
elements=driver.find_elements(By.CSS_SELECTOR, 'td[class="sticky z-10 left-0 pl-2.5 sm:pl-3.75 bg-white py-2.5"]')
for element in range(len(elements)):
    date.append(elements[element].text)


print(project)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(funding_rounds)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(total_funding_amount)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(date)
print('-----------------------------------------------------------------------------------------------------------------------------------')
print(len(project),len(funding_rounds),len(total_funding_amount),len(date))
```
