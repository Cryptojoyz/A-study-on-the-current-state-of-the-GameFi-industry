```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
```

```python
browser = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
browser.implicitly_wait(10)
browser.get("https://chainplay.gg/guild-performance/")
```

```python
guild=[]
price=[]
market_cap=[]
discord=[]
partnered_games=[]
investors=[]
# 工会名称
elements=browser.find_elements(By.CSS_SELECTOR,".capitalize")
for element in elements:
    guild.append(element.text.strip())
# 代币价格
elements=browser.find_elements(By.CSS_SELECTOR,"table .border-b p")
for element in elements:
    price.append(element.text.strip())
# 市值
elements=browser.find_elements(By.CSS_SELECTOR,'tr.border-b td:nth-of-type(4) ')
for element in elements:
    market_cap.append(element.text.strip())
# discord软件指数
elements=browser.find_elements(By.CSS_SELECTOR,'tr.border-b td:nth-of-type(6) ')
for element in elements:
    discord.append(element.text.strip())
# 合作游戏数量
elements=browser.find_elements(By.CSS_SELECTOR,'tr.border-b td:nth-of-type(7) ')
for element in elements:
    partnered_games.append(element.text.strip())
# 展开
elements=browser.find_elements(By.CSS_SELECTOR,'tr.border-b td:nth-of-type(8) button')
for element in elements:
    element.click()
#投资者
elements=browser.find_elements(By.CSS_SELECTOR,'tr.border-b td:nth-of-type(8) ')
for element in elements:
    investors.append(element.text)

browser.quit()
```

```python
print(len(guild))
print(len(price))
print(len(market_cap))
print(len(discord))
print(len(partnered_games))
print(len(investors))
```

```python
import pandas as pd
data_guilds={
    "guilds name":guild,
    "token price":price,
    "market cap":market_cap,
    "discord":discord,
    "number of partnered games":partnered_games,
    "investors":investors
}

pd_data_guilds = pd.DataFrame.from_dict(data_guilds)
pd_data_guilds.index=pd_data_guilds.index+1
pd_data_guilds
```

```python
pd_data_guilds.to_csv('data_guilds.csv')
```
