```python
from bs4 import BeautifulSoup
import pandas as pd
import time
from time import sleep
from random import randint
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('http://www.chainplay.gg/investors/')
```

```python
name=[]
investor_type=[]
total_investment=[]
current_roi=[]
ath_roi=[]
top3_roi=[]


page=driver.page_source
soup=BeautifulSoup(page,"lxml")
b=soup.select('div[class="flex-1 overflow-hidden"]')
for data in b:
    j=[]
    for i in data.find_all('a'):
        j.append(i['href'])
    name.append(j)  


elements = driver.find_elements(By.CSS_SELECTOR, 'table tbody div[class="flex-1 overflow-hidden"] small')
for element in range(len(elements)):
    investor_type.append(elements[element].text)
        

elements = driver.find_elements(By.CSS_SELECTOR, 'td[class="px-2.5 sm:px-3.75 py-2.5 text-right"] ')
for element in range(len(elements)):
    total_investment.append(elements[element].text)        



elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(4) ')
for element in range(len(elements)):
    current_roi.append(elements[element].text)
    

elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(5) ')
for element in range(len(elements)):
    ath_roi.append(elements[element].text)

page=driver.page_source
soup=BeautifulSoup(page,"lxml")
b=soup.select('td[class="sm:px-3.75 py-2.5"]')
for data in b:
    j=[]
    for i in data.find_all('a'):
        j.append(i.text)
    top3_roi.append(j) 



print(name)
print('---------------------------------------------------------------------------------------------------------------------------------')
print(investor_type)
print('---------------------------------------------------------------------------------------------------------------------------------')
print(total_investment)
print('---------------------------------------------------------------------------------------------------------------------------------')
print(current_roi)
print('---------------------------------------------------------------------------------------------------------------------------------')
print(ath_roi)
print('---------------------------------------------------------------------------------------------------------------------------------')
print(top3_roi)
print('---------------------------------------------------------------------------------------------------------------------------------')
print(len(name),len(investor_type),len(total_investment),len(current_roi),len(ath_roi),len(top3_roi))
```

```python
for m in range(13):
    time.sleep(0.5)
    
    button= driver.find_element(By.CSS_SELECTOR, 'button[class="w-8 h-8 flex items-center justify-center rounded-lg hover:bg-[#edf2f7] "]:nth-last-child(1)')
    button.click()
    
    time.sleep(1)
    
    page=driver.page_source
    soup=BeautifulSoup(page,"lxml")
    b=soup.select('div[class="flex-1 overflow-hidden"]')
    for data in b:
        j=[]
        for i in data.find_all('a'):
            j.append(i['href'])
        name.append(j)  


    elements = driver.find_elements(By.CSS_SELECTOR, 'table tbody div[class="flex-1 overflow-hidden"] small')
    for element in range(len(elements)):
        investor_type.append(elements[element].text)
        

    elements = driver.find_elements(By.CSS_SELECTOR, 'td[class="px-2.5 sm:px-3.75 py-2.5 text-right"] ')
    for element in range(len(elements)):
        total_investment.append(elements[element].text)        
        

    elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(4) ')
    for element in range(len(elements)):
        current_roi.append(elements[element].text)
    

    elements = driver.find_elements(By.CSS_SELECTOR, 'tr[class="border-b hover:bg-gray-100"] td:nth-of-type(5) ')
    for element in range(len(elements)):
        ath_roi.append(elements[element].text)

    page=driver.page_source
    soup=BeautifulSoup(page,"lxml")
    b=soup.select('td[class="sm:px-3.75 py-2.5"]')
    for data in b:
        j=[]
        for i in data.find_all('a'):
            j.append(i.text)
        top3_roi.append(j)     

print(len(name),len(investor_type),len(total_investment),len(current_roi),len(ath_roi),len(top3_roi))
```

```python
investor_name=[]
for a in name:
    inner=[]
    for b in a:
        inner.append(b.replace('/investor/',''))
    investor_name.append(inner)
print(investor_name)
```

```python
name1=[]
for a in investor_name:
    b=",".join(a)
    name1.append(b)
print(name1)
```

```python
top3_roi1=[]
for a in top3_roi:
    b=",".join(a)
    top3_roi1.append(b)
print(top3_roi1)
```

```python
import pandas as pd
tiny={"Name":name1,
      "Investor Type":investor_type,
      "Total Investment":total_investment,
      "Current ROI":current_roi,
      "Ath ROI":ath_roi,
      "Top3 ROI":top3_roi1
     }

df=pd.DataFrame(tiny)
df.index=df.index + 1
df
```
